```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_circles

# Example 1: Two interleaving half circles (nonlinear)
# X, y = make_moons(n_samples=500, noise=0.1, random_state=42)

# Example 2: Concentric circles (nonlinear)
# X, y = make_circles(n_samples=500, noise=0.05, factor=0.5, random_state=42)

# Example 3: Custom nonlinear transformation
# Random points
X = np.random.uniform(-1, 1, (500, 2))
y = (X[:, 0]**2 + X[:, 1]**2 > 0.5).astype(int)  # Circle boundary

# Visualize
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolor="k")
plt.title("Nonlinear Dataset")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples
n_samples = 300

# Generate random points
X = np.random.uniform(-0.6, 0.6, (n_samples, 2))

# Nonlinear decision boundary
# Here: classify based on x^2 < y  (you can tweak this function)
y = (X[:, 1] > X[:, 0]**2).astype(int)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.plasma, edgecolor="k")
plt.title("Custom Nonlinear Dataset")
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense ,Input
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

```python
model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(128, activation = "linear"))
model.add(Dense(128, activation = "linear"))
model.add(Dense(1, activation = "sigmoid"))


adam =Adam(learning_rate = 0.01)
model.compile(loss= "binary_crossentropy", optimizer = adam , metrics = ['accuracy'])


history = model.fit(X,y , epochs = 500 , validation_split = 0.2 , verbose = 0 )
```

```python
model.summary()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf = model , legend =2)
plt.xlim(-0.7,0.5)
plt.ylim(-0.8,0.8)
plt.show()
```
